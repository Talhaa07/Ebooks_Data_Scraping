```python
title = []
publisher = []
subtitle = []
publication_year = []
auther = []
price = []
description = []
for books in book:
  title.append(books['title'])
  subtitle.append(books['subtitle'])
  publisher.append(books['publisher'])
  publication_year.append(books['publication_year'])
  try:
    auther.append(books['authors'][0]['name'])
  except:
    None
  price.append(books['price'])
  description.append(books['description'])
```

```python
df = pd.DataFrame({'Title':title,"Publisher":publisher,"auther":auther,"publication":publication_year,"price":price,"descriptions":description,'subtitle':subtitle})
```
